### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

1. pyplot
2. Objekt Orienterat (OO)

Pyplot är mer övergripande, mindre detaljer (egenskaper) att fylla ut. OO kan man få tillgång till fler egenskaper och därigenom få bättre kontroll över objektens egenskaper och återanvändning, därför är OO metoden att föredra. 

```python
import matplotlib.pyplot as plt

plt.plot([1.5, 2, 3.8], [0.7, 1.1, 2.4], color='red', marker='o', linestyle='--')
plt.xlabel('Motorstorlek')
plt.ylabel('Förbrukning')
plt.title('Förbrukning vs motorstorlek')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([1.5, 2, 3.8], [0.7, 1.1, 2.4], color='blue', marker='o', linestyle=':')
ax.set_xlabel('Motorstorlek')
ax.set_ylabel('Förbrukning')
ax.set_title('Förbrukning vs motorstorlek')
ax.grid(True)
plt.show()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

figure - själva bakgrunden (canvas)
artist - allt som är synligt på figure (text, x,y-axlar, statiska texter, etc)
axes - själva graf-plot-ytan 
axis - kontrollera x,y,z -axlarnas egenskaper med tjocklek, skalstreck, texter, mm

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-4, 4, 50)
y = x**2

fig, ax = plt.subplots()
ax.set_facecolor('lightgray')
ax.plot(x, y, color='green', marker='', linestyle='-')
ax.set_xlabel('X-axel')
ax.set_ylabel('Y-axel')
ax.set_title('OO-Graf y = x^2')
ax.grid(True)
plt.show()

plt.plot(x, y)
plt.gca().set_facecolor('lightgray')
plt.title('Plot y = x^2')
plt.xlabel('X-axel')
plt.ylabel('Y-axel')
plt.grid(True)
plt.show()

```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt 

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000) # Generate 1000 random numbers for x-axis
random_data_y = np.random.randn(1000) # Generate 1000 random numbers for y-axis

# Scatter plot of random data
fig, ax = plt.subplots() 
ax.scatter(random_data_x, random_data_y, color='blue', alpha=0.5, marker='x')
ax.set_facecolor('lightyellow')
ax.set_title('Slumpade punkter')
ax.set_xlabel('X-axel')
ax.set_ylabel('Y-axel')
ax.grid(True)
plt.show()  

# ------------------- X -------------------

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Bar plot of fruit counts
fig, ax = plt.subplots()
ax.bar(names, values, color='orange', alpha=0.7, edgecolor='black')
ax.set_facecolor('lightblue')
ax.set_title('Fruits Count')            
ax.set_xlabel('Fruit')
ax.set_ylabel('Count')
ax.grid(axis='y')
plt.show()
```
