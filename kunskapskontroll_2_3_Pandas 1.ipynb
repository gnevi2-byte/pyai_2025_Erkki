### Explain what a CSV file is.

Comma Separated Values (CSV) is a common data format used for storing tabular data, 
such as spreadsheets or databases. Each line in a CSV file represents a row of data, 
and each value within that row is separated by a comma. CSV files are widely used because they are simple, human-readable, and can be easily imported and exported by various software applications, including spreadsheet programs like Microsoft Excel and data analysis tools like Python's Pandas library.

En CSV är helt enkelt en textfil där datat är separerade med ett komma tecken. CSV filer är väldigt mobila i o m att det är en enkel textfil. De flesta seriösa program (av databas karaktär) kan exportera och importera csv filer. T ex exportera ut en signalista från ett scada system för att sedan efter manipulering importera in den i ett nytt system. Verktyg för manipulation brukar vara excel. Excel är ett bra program för att hantera data i tabular form.

T ex på innehåll i en CSV fil;

Index, Namn, telefon, ....
1, Erkki, 021212121, ....
2, Kalle, 081232121, ....
3, Kajsa, 031784125, ....
...

### Print the first 10 rows of the data. 

```python
import pandas as pd
from io import StringIO

f =  pd.read_csv(r'cars_data.csv')   
print(f.head(10)) # Display first 10 rows of the dataframe
```

### Print the last 5 rows. 

```python
import pandas as pd
from io import StringIO

f =  pd.read_csv(r'cars_data.csv')   
print("5 last rows:\n", f.tail(5)) # Display last 5 rows of the dataframe
```

### By using the info method, check how many non-null rows each column have. 

```python
print(f.info()) # Get info about the dataframe including null values. 
print()
print(f.isnull().sum()) # Check for null values in each column. (omvänt mot ovan, visar antal rader med null värden)

#print(f.describe()) # Get statistical summary of numerical columns.
```

I första fallet ser vi att 'price' avviker med 3, i andra fallet (isnull) så visas också att 'price' 3 värden (är null).

### If any column has a missing value, drop the entire row. Notice, the operation should be inplace meaning you change the dataframe itself.

```python
f.dropna(inplace=True)  # Remove rows with null values ( idetta fall de tre raderna med saknade värden, verifierat med excel )

print(f.info()) # Get info about the dataframe including null values.
print()
print(f.isnull().sum()) # Check for null values in each column. (omvänt mot ovan, visar antal rader med null värden)
```

### Calculate the mean of each numeric column. 

```python
print(f.describe()) # Get statistical summary of numerical columns.
```

### Select the rows where the column "company" is equal to 'honda'. 

```python
#print(f.info())

f_df = f[f['company'] == 'honda']  # Filter rows where company is 'honda'
print(f_df)
```

### Sort the data set by price in descending order. This should *not* be an inplace operation. 

```python
s_df = f.sort_values(by='price', ascending=False, inplace=False).head(5)  # Get top 5 most expensive cars
print(s_df)
```

### Select the rows where the column "company" is equal to any of the values (audi, bmw, porsche).

```python
f_df = f[f['company'].isin(['audi', 'bmw', 'porsche' ])]  # Filter rows where company is 'audi', 'bmw', or 'porsche'
print(f_df)
```

### Find the number of cars (rows) for each company. 

```python
f_noc = f['company'].value_counts()  # Count number of cars per company
print(f_noc)
```

### Find the maximum price for each company. 

```python
maxp = f.groupby('company')['price'].max()  # Get maximum price per company
minp = f.groupby('company')['price'].min()  # Get minimum price per company
medp = f.groupby('company')['price'].median()  # Get median price per company

print("Maximum price:\n", maxp)
print("\nMinimum price:\n", minp)
print("\nMedian price:\n", medp)
```
